```python
df['Frequency_Salary'] = df['salary'].str.split(' a ').str.get(1)
```

```python
df['Frequency_Salary'].unique()
```

```python
df.head()
```

```python
salary_split = df['salary'].str.split('-', expand=True)
df['Low_Salary'] = salary_split[0].str.strip()
df['High_Salary'] = salary_split[1].str.strip()
```

```python
df.head()
```

```python
patterns_to_remove = r'[\₹,\$]|(month|months)|per hour|(year|years)|a|from|From\b'
df['Low_Salary'] = df['Low_Salary'].replace(patterns_to_remove, '', regex=True)
df['High_Salary'] = df['High_Salary'].replace(patterns_to_remove, '', regex=True)
```

```python
df.head()
```

```python
df['Low_Salary'] = pd.to_numeric(df['Low_Salary'], errors='coerce')
df['High_Salary'] = pd.to_numeric(df['High_Salary'], errors='coerce')
```

```python
df.head()
```

```python
def transform_salary(row):
    if row['Frequency_Salary'] == 'year':
        factor = 1
    elif row['Frequency_Salary'] == 'month':
        factor = 12
    elif row['Frequency_Salary'] == 'week':
        factor = 49
    elif row['Frequency_Salary'] == 'day':
        factor = 230
    elif row['Frequency_Salary'] == 'hour':
        factor = 1840
    else:
        factor = 1
    row['Low_Salary'] *= factor
    row['High_Salary'] *= factor
    return row

df = df.apply(transform_salary, axis=1)
```

```python
df.head()
```

```python
df['total_salary'] = (df['Low_Salary'] + df['High_Salary']) / 2
```

```python
df.head()
```

```python
df['total_salary'] = df['total_salary'].fillna('Not Applicable')
```

```python
location_split = df['company_location'].str.split(',', expand=True)
df['City'] = location_split[0].str.strip()
df['State'] = location_split[1].str.strip()
```

```python
df.head()
```

```python
df['company_location'] = df['company_location'].str.replace('Hybrid remote in ', '', regex=False)
df['company_location'] = df['company_location'].str.replace('Remote in ', '', regex=False)
```

```python
df.head()
```

```python
salary_split = df['company_location'].str.split(',', expand=True)
df['City'] = salary_split[0].str.strip()
df['State'] = salary_split[1].str.strip()
```
